## Initializing Llamma 2 model and the corresponding Tokenizer

```python
# Initialize the Llama 2 model and tokenizer
def load_llama2_model():
    model_name = "meta-llama/Llama-2-7b-chat-hf"

    # Load tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    # Load model
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        device_map="auto",
        torch_dtype=torch.float16,
    )

    # Create text generation pipeline
    pipe = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
        max_new_tokens=512,
        temperature=0.1, # remember for anlysis: Lower temprature is agent-freindly
        top_p=0.95,
        repetition_penalty=1.15,
        return_full_text=False,

    )

    # Wrap in LangChain HuggingFacePipeline
    llm = HuggingFacePipeline(pipeline=pipe)

    return llm

llm = load_llama2_model()
```

#### Teacher Assistant class

```python
class TeacherAssistant:

    def __init__(self):
        # Define triggers
        self.react_triggers = [

            r"verify", r"fact check", r"according to",
            r"when did", r"who invented", r"what is the",
            r"current", r"2024", r"2023"
        ]
        self.few_shot_triggers = [
            r"student", r"classroom", r"noisy",
            r"attention", r"how to handle"
        ]

        self.classroom_examples = [
            {"situation": "student struggling", "response": "Let's break it down step by step."},
            {"situation": "class is noisy", "response": "Eyes on me in 3...2...1. Thank you."}
            ]

    def define_tools(self):

        # The repetition here is beacause of the parsing error I keep geting when using
        # agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION
          wikipedia = WikipediaAPIWrapper()
          wikipedia_tools = tools = [
                Tool(
                    name="Search",
                    func=wikipedia.run,
                    description="Useful for searching general knowledge on Wikipedia"
                ),
                Tool(
                    name="Lookup",
                    func=wikipedia.run,
                    description="Useful for looking up specific facts on Wikipedia"
                )
            ]

          return wikipedia_tools

    def use_ReAct(self, query):

        """This function will be called when we want to use ReAct Prompting"""

        print("Using REACT + Wikipedia")
        wikipedia_tools = self.define_tools()

        agent = initialize_agent(
                tools = wikipedia_tools,
                llm=llm,
                agent = AgentType.REACT_DOCSTORE,
                verbose = True,
                handle_parsing_errors = True,
                max_iterations = 2,
                # early_stopping_method = "generate"
            )

        print("Created an agent")
        react_query = f"As a knowledgeable teaching assistant, use the tools to answer the following academic question factually and concisely: {query}"
        print(f"Running agent with query: {react_query}")
        response = agent.run(react_query)

        return response

    def use_FewShot(self, query):
        """This function will be called when we want to use Few_shot Prompting"""
        print("Using FEW-SHOT examples")

        prompt = FewShotPromptTemplate(
            examples = self.classroom_examples,
            example_prompt = PromptTemplate(
                input_variables = ["situation", "response"],
                template = "Situation: {situation}\nResponse: {response}"
            ),
            prefix = "As an expert teacher, respond appropriately:",
            suffix = "Situation: {input}\nResponse:",
            input_variables = ["input"]
        )
        return llm(prompt.format(input=query))

    def use_ZeroShot(self, query):
        """This function will be called when we want to use Zero_shot prompting"""

        print("Using ZERO-SHOT")
        return llm(f"As a teaching assistant, answer concisely: {query}")



    def route(self, query: str) -> str:
        query_lower = query.lower()

        # REACT: Fact verification and Real-Time Information
        if any(re.search(pattern, query_lower) for pattern in self.react_triggers):

            return self.use_ReAct(query)


        # FEW-SHOT: Classroom scenarios
        elif any(re.search(pattern, query_lower) for pattern in self.few_shot_triggers):

            return self.use_FewShot(query)

        # ZERO-SHOT: Everything else
        else:

            return self.use_ZeroShot(query)

```

### Execution Cell

```python
assistant = TeacherAssistant()

test_queries = [

    "How to handle noisy students during exams",   # Few-Shot
    "Verify if photosynthesis requires sunlight",  # ReAct
    "Who is the current president of Ethiopia?",   # ReAct
    "A student is crying in the hallway",          # Few-Shot
    "Explain quantum computing to 5th graders",     # Zero-Shot

    # Edge cases
    "verify when shanon's noisy theorem was proposed"
    # hell ya, It passed the edge case
]

for query in test_queries:
    print(f"\n{'='*60}")
    print(f"QUERY: {query}")
    response = assistant.route(query)
    print(f"RESPONSE: {response}")
    print('='*60)
```
